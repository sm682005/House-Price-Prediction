```python
##Independent and Dependent Features
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
###Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=97)
```

```python
##  Standardize the dataset 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
x_test
```

# Model Training 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
## print the coefficients  and the intercept 
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
## On which parameter the model has been trained 
lr.get_params()
```

```python
###Prediction woth test data
reg_pred = lr.predict(x_test)
```

```python
reg_pred
```

```python
### plot a scatter plot for the prediction 
plt.scatter(y_test,reg_pred)
plt.show()
```

```python
#when we see both the values predicted and actual output value are linear in nature using scatter plot basically shown our model is completly done 
#and predicting well.
```

```python
# residual means error between the predicted value and actual ouput value (difference between both the outputs)
residuals = y_test - reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
plt.show()
```

```python
##scatter plot with repect to predictions asnd residuals
##uniform distribution
plt.scatter(reg_pred,residuals)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```
